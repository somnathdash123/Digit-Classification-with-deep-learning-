```python

import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
y_train.shape
```

```python
X=X_train/255
x=X_test/255
```

```python
Xf= X.reshape(len(X), 28*28)
xf = x.reshape(len(x), 28*28)
```

```python

model = keras.Sequential([
    keras.layers.Dense(100, input_shape=( 28*28, ), activation='relu'), 
    keras.layers.Dense(50,activation='relu' ),
    keras.layers.Dense(10,activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(Xf, y_train, epochs=10)
```

```python
def pre (x):
    x=x.reshape(1, 28*28)
    yy=model.predict(x)
    return np.argmax(yy)
```

```python
def check (x):
    print('True answer :', x)
```

```python

model.evaluate(xf, y_test)
```

```python
n=int(input("Enter the index between (0 - {}):".format(len(y_test))))
print("Predicted answer :", pre(xf[n]))
check(y_test[n])
```
